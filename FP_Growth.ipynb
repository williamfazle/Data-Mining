```python
import pandas as pd
from collections import Counter

# ---------- DATASET ----------
data = {
    "Item1": ["milk", "bread", "eggs", "juice", "milk"],
    "Item2": ["bread", "butter", "nan", "chips", "nan"],
    "Item3": ["eggs", "nan", "nan", "nan", "tea"]
}

df = pd.DataFrame(data)

# ---------- PREPROCESSING ----------
transactions = []

for i in range(df.shape[0]):
    row_items = []

    for j in range(df.shape[1]):
        item = df.values[i, j]

        # (1) Fill in the line to skip "nan" values
        if item == "nan":
            continue

        # (2) Fill in the line to append valid items to row_items
        row_items.append(item)

    # (3) Add each completed row_items list into transactions
    transactions.append(row_items)

# ---------- CALCULATE SUPPORT ----------
item_counts = Counter()

# (4) Fill in the line that updates the item_counts Counter
for trans in transactions:
    item_counts.update(trans)

total_transactions = len(transactions)
supports = {item: count/total_transactions for item, count in item_counts.items()}

print("Transactions:")
print(transactions)

print("\nItem Supports:")
print(supports)
```
